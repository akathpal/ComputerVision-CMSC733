```python
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
from scipy.optimize import minimize
%matplotlib inline
```

```python
def estimateParams(objp,corners):
    v = []
    H1,_ = cv2.findHomography(objp,corners)
    H = H1.transpose()
    list = {(0,1),(0,0),(1,1)}
    for i,j in list:
        v.append([H[i,0]*H[j,0],H[i,0]*H[j,1]+ H[i,1]*H[j,0],H[i,1]*H[j,1],H[i,2]*H[j,0]+H[i,0]*H[j,2],H[i,2]*H[j,1]+H[i,1]*H[j,2],H[i,2]*H[j,2]])

    r1 = v[0]
    r2 = np.array(v[1])-np.array(v[2])

    
    return r1,r2,H1

    
```

```python
# termination criteria
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)

out = np.zeros(shape=(26,6))
# prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(6,5,0)
objp = np.zeros((6*9,3), np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

# Arrays to store object points and image points from all the images.
objpoints = [] # 3d point in real world space
imgpoints = [] # 2d points in image plane.
homography =[]
images = glob.glob('Calibration_Imgs/*.jpg')
# print(len(images))
# out = []
i=0
for fname in images:
    img = cv2.imread(fname)
#     print(fname)
#     gray = img
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    # Find the chess board corners
    ret, corners = cv2.findChessboardCorners(gray, (9,6),None)
#     print(corners.shape)
    
    

    # If found, add object points, image points (after refining them)
    if ret == True:
        objpoints.append(objp)
        corners=corners.reshape(-1,2)
        corners2 = cv2.cornerSubPix(gray,corners,(11,11),(-1,-1),criteria)
        imgpoints.append(corners2)

        r1,r2,H = estimateParams(objp,corners2)
        homography.append(H)
        
        out[i] = r1
        out[i+1] = r2
        i = i+2
        
        
print(out)
# print(len(homography))
#         # Draw and display the corners
#         img = cv2.drawChessboardCorners(img, (7,6), corners2,ret)
# #         cv2.imshow('img',img)
#         plt.figure(figsize=(15,15))
#         plt.imshow(img)
#         cv2.waitKey(500)


    
    
```
